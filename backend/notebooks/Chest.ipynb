```python
# training
import torch
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
```

```python
BATCH_SIZE = 16
EPOCHS = 7
LEARNING_RATE = 1e-3
NO_LABELS = 20
IMG_PATH = '../data/images/'
```

```python
from PIL import Image
import torchvision.transforms as T
import cv2
```

```python
# Setting up the device for GPU usage
from torch import cuda
device = 'cuda' if cuda.is_available() else 'cpu'
print(device)
```

```python
transforms = T.Compose([
                       T.ToPILImage(), 
                       T.Resize((299, 299)),
                       T.ToTensor(),
                       T.Normalize(
                                   mean=[0.485, 0.456, 0.406],
                                   std=[0.229, 0.224, 0.225]
                                   )
                      ])
```

```python
# Creating Dataset and DataLoader for neural net
class ChestDataset(Dataset):

    def __init__(self, dataframe, transforms, IMG_PATH):
        self.transform = transforms
        self.data = dataframe
        self.image_path = IMG_PATH
        self.image_names = self.data.id
        self.labels = self.data.labels

    def __len__(self):
        return len(self.image_names)

    def __getitem__(self, index):
        image = cv2.imread(f'{self.image_path}/{self.image_names[index]}')
        image = self.transform(image)
        targets = torch.tensor(self.labels[index], dtype=torch.float)

        return {
            'image': image,
            'labels': targets
        }
```

```python
train_set = ChestDataset(train, transforms, IMG_PATH)
test_set = ChestDataset(test, transforms, IMG_PATH)
```

```python
train_params = {'batch_size': BATCH_SIZE,
                'shuffle': True,
                'num_workers': 0
               }

test_params = {'batch_size': BATCH_SIZE,
              'shuffle': True,
              'num_workers': 0
             }

train_loader = DataLoader(train_set, **train_params)
test_loader = DataLoader(test_set, **test_params)
```

```python
# creating and fine tuning customized model on top of BERT
import torch
from torch import nn
from torchvision import models

class ChestModel(torch.nn.Module):
    def __init__(self, no_labels):
        super(ChestModel, self).__init__()
        model = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
        # freezing weights for transfer learning
        #for param in model.parameters():
        #    param.requires_grad = False  
        
        model.fc = nn.Sequential(
            nn.Dropout(p=0.2),
            nn.Linear(in_features=model.fc.in_features, out_features=no_labels)
        )
        self.base_model = model

    
    def forward(self, image):
        outputs = self.base_model(image)
        
        return outputs
```

```python
model = ChestModel(NO_LABELS)
model.to(device)
```

```python
# training function
def train(model, dataloader, optimizer, criterion, acc_fn, device, best_acc):
    model.train()
    loop = tqdm(dataloader)
    train_loss = 0
    train_acc = 0
    for i, data in enumerate(loop):
        
        image, target = data['image'].to(device), data['labels'].to(device, dtype = torch.float)
        
        optimizer.zero_grad()
        
        outputs = model(image)
        
        loss = criterion(outputs, target)
        acc = acc_fn(outputs, target)
        
        train_loss += loss.item()
        train_acc += acc.item()
        
        # progress bar
        loop.set_description("Train")
        loop.set_postfix(train_loss=loss.item(), train_accuracy=acc.item())
        
        # backpropagation
        loss.backward()
        # update optimizer parameters
        optimizer.step()
    
    loop.set_postfix(train_loss=train_loss/len(dataloader), train_accuracy=train_acc/len(dataloader))
    
    if train_acc > best_acc:
        best_acc = train_acc
        PATH = 'chest_model1.pth'
        torch.save(model.state_dict(), PATH)
    
    return train_loss, train_acc, best_acc
```

```python
# test function
def test(model, dataloader, criterion, acc_fn, device):
    model.eval()
    loop = tqdm(dataloader)
    test_loss = 0
    test_acc = 0
    with torch.no_grad():
        for i, data in enumerate(loop):
            
            image, target = data['image'].to(device), data['labels'].to(device, dtype = torch.float)
            outputs = model(image)
            
            loss = criterion(outputs, target)
            acc = acc_fn(outputs, target)
            
            test_loss += loss.item()
            test_acc += acc.item()
            
            # progress bar
            loop.set_description("Test")
            loop.set_postfix(testing_loss=loss.item(), test_accuracy=acc.item())
            
        loop.set_postfix(testing_loss=test_loss/len(dataloader), test_accuracy=test_acc/len(dataloader))

    return test_loss, test_acc
```

```python
# defining loss function
def loss_fn(outputs, targets):
    return torch.nn.CrossEntropyLoss()(outputs, targets)
```

```python
def multi_acc(y_pred, y_test):
    y_pred_softmax = torch.log_softmax(y_pred, dim = 1)
    _, y_pred_tags = torch.max(y_pred_softmax, dim = 1)
    y_test = torch.tensor([(y.index(1)) for y in y_test.tolist()]).to(device)
   
    correct_pred = (y_pred_tags == y_test).float()
    acc = correct_pred.sum() / len(correct_pred)
    
    acc = torch.round(acc * 100)
    
    return acc
```

```python
# setting optimizer
optimizer = torch.optim.Adam(params=model.parameters(), lr=LEARNING_RATE)
```

```python
# start the training and testing
train_loss = []
train_acc = []

test_loss = []
test_acc = []
best_acc = 0
for epoch in range(EPOCHS):
    print(f"Epoch {epoch+1}/{EPOCHS}:")
    train_epoch = train(model, train_loader, optimizer, loss_fn, multi_acc, device, best_acc)
    test_epoch = test(model, test_loader, loss_fn, multi_acc, device)
    
    train_loss.append(train_epoch[0])
    train_acc.append(train_epoch[1])
    
    test_loss.append(test_epoch[0])
    test_acc.append(test_epoch[1])
    
    best_acc = train_epoch[2]
```

```python
# saving the trained model for inference
PATH = 'chest_model_resnet50.pth'
torch.save(model.state_dict(), PATH)
```

```python
# creating and fine tuning customized model on top of BERT
import torch
from torch import nn
from torchvision import models

class ChestModel(torch.nn.Module):
    def __init__(self, no_labels):
        super(ChestModel, self).__init__()
        model = models.vgg16(weights='DEFAULT')
        # freezing weights for transfer learning
        for param in model.parameters():
            param.requires_grad = False  
        
        n_inputs = model.classifier[6].in_features
        # n_inputs will be 4096 for this case
        # Add on classifier
        model.classifier[6] = nn.Sequential(
                                    nn.Linear(n_inputs, 256),
                                    nn.ReLU(),
                                    nn.Dropout(0.4),
                                    nn.Linear(256, no_labels))
        self.base_model = model

    
    def forward(self, image):
        outputs = self.base_model(image)
        
        return outputs
```
